```python
from datetime import datetime, timedelta

# Define the start and end dates
start_date = datetime.strptime('02/12/2022', '%d/%m/%Y')
end_date = datetime.strptime('24/04/2024', '%d/%m/%Y')

# Initialize a counter for Sundays
sundays_count = 0

# Iterate over each date in the range
current_date = start_date
while current_date <= end_date:
    # Check if the current date is a Sunday (weekday() returns 6 for Sunday)
    if current_date.weekday() == 6:
        sundays_count += 1
    # Move to the next day
    current_date += timedelta(days=1)

# Print the number of Sundays
print("Number of Sundays between", start_date.strftime('%d/%m/%Y'), "and", end_date.strftime('%d/%m/%Y'), ":", sundays_count)
```

```python
import re
from datetime import datetime

# Read the text file
with open('WhatsApp_Chat_with_FLA_Ushers_City.txt', 'r') as file:
    data = file.readlines()

# Define a regular expression pattern to match date formats
date_pattern = r'(\d{2}/\d{2}/\d{4})'

# Function to check if a given date is a Sunday
def is_sunday(date_str):
    date_obj = datetime.strptime(date_str, '%d/%m/%Y')
    return date_obj.weekday() == 6  # 6 represents Sunday

# Filter out messages sent on Sundays
sunday_messages = [line.strip() for line in data if re.search(date_pattern, line) and is_sunday(re.search(date_pattern, line).group(1))]

# Print the filtered messages
for message in sunday_messages:
    print(message)
```

```python
# Define the output file path
output_file_path = 'sunday_messages.txt'

# Write the filtered messages to the output file
with open(output_file_path, 'w') as output_file:
    for message in sunday_messages:
        output_file.write(message + '\n')

print(f"Filtered messages saved to {output_file_path}")
```

```python
import re

# Read the text file
with open('sunday_messages_counts_filtered.txt', 'r') as file:
    data = file.read()

# Define a regular expression pattern to match the counts
count_pattern = r'(?:(?:First|Second|Final) count|Final seated count):? (\d+)'

# Find all matches of the pattern in the text
counts = re.findall(count_pattern, data)

# Create a list to store the filtered counts
filtered_counts = []

# Iterate through the matches and append "NA" for missing counts
for count in counts:
    if count.isdigit():
        filtered_counts.append(int(count))
    else:
        filtered_counts.append("NA")

# Print the filtered counts
print(filtered_counts)

# Write the filtered counts to a file
with open('filtered_counts.txt', 'w') as file:
    for count in filtered_counts:
        file.write(str(count) + '\n')
```

```python
import matplotlib.pyplot as plt

# Read the filtered counts from the file
with open('filtered_counts.txt', 'r') as file:
    filtered_counts = [int(line.strip()) if line.strip().isdigit() else None for line in file]

# Generate x-axis labels (indices of counts)
x_labels = range(1, len(filtered_counts) + 1)

# Create the plot
plt.figure(figsize=(10, 6))
plt.bar(x_labels, filtered_counts, color='skyblue')

# Set plot title and labels
plt.title('Attendance Counts Over Time')
plt.xlabel('Weeks')
plt.ylabel('Attendance Count')

# Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
